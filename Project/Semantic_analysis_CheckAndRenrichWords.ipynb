```python
import jieba, codecs, word2vec, gensim, logging, scipy, uniout
from gensim.models import Word2Vec
```

```python
'''
讀取儲存的訓練結果，使用word讀取的字詞，逐一做字詞的相似度檢查。
並且利用check這個Dict做檢查，是否有重複。
'''
def word2VecDictCheck(load_model, savePath, checkPath):
    check = dict()
    keywordInput = dict()
    with open(savePath, 'w') as wordsWrite:
        for words in open(checkPath, 'r'):
            check[words] = words
        for words in check:
            try:
                for keyword, weight in load_model.most_similar(u'%s' % words.decode('utf-8'), topn = 1000):
                    if keyword not in check:
                        keywordInput[keyword] = keyword
                        wordsWrite.write(keywordInput[keyword].encode('utf-8') + '\n')
                    else:
                        print keyword
            except:
                print 'KeyError happened!!'
                continue
```

```python
'''
這邊是做一般的字詞檢查，將為重複的放入keywordInput的Dict內，
並且寫入到檔案中。
'''
def normalDictCheck(savePath, readPath, checkPath):
    check = dict()
    keywordInput = dict()
    with open(savePath, 'w') as wordsWrite:
        for words in open(checkPath, 'r'):
            check[words] = words
        for keyword in open(readPath, 'r'):
                if keyword not in check:
                    keywordInput[keyword] = keyword.decode('utf-8')
                    wordsWrite.write(keywordInput[keyword].encode('utf-8'))
                else:
                    print keyword
```
