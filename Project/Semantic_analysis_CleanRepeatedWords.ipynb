```python
import codecs

def checkAndDelete(filePath):
    with open(filePath, 'r') as dataRead:
        check = dict()
        print '-' * 10 + 'Repeated Words as below' + '-' * 10
        for lines in dataRead.readlines():
            if lines.strip() not in check:
                check[lines.strip()] = lines.strip()
            else:
                print lines.strip()
        return check

def wrtieToFile(savePath, check):
    with open(savePath, 'w') as dataWrite:
        for words in check:
            dataWrite.write(words + '\n')
            
def start(filePath, savePath, fileName):
    newSavePath = savePath + '\%s.txt' % fileName
    check = checkAndDelete(filePath)
    wrtieToFile(newSavePath, check)
    print '-' * 10 + 'All are done!' + '-' * 10
```
